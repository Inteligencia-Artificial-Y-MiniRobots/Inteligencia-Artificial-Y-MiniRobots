# Reglas Basicas De Comportamiento

## Reglas Básicas para el Comportamiento en la Casa

- Limpieza y Orden: Manténer el espacio limpio y ordenado.

- Respeto por la Privacidad: Respetar la privacidad de los demás habitantes de la casa.

- Colaboración en las Tareas del Hogar: Contribuir con las tareas domésticas compartidas, como la limpieza, la cocina y el mantenimiento.

- Seguridad: Tomar medidas para mantener la casa segura.

- Comunicación Abierta: Comunicar las necesidades y preocupaciones de manera abierta y respetuosa.

- Uso Responsable de Recursos: Usar los recursos de la casa, como el agua y la electricidad, de manera responsable.

- Cuidado de las Mascotas : Asumir la responsabilidad de cuidarlas según las necesidades, como alimentarlas y sacarlas a pasear.

- Cuidado de las Áreas Comunes: Mantener limpias y en buen estado las áreas comunes de la casa, como la sala de estar y la cocina.

## Reglas Básicas para el Comportamiento en la Universidad

- Asistencia y Puntualidad: Asistor a clases de manera regular y llegar a tiempo.

- Respeto Académico: Respetar los derechos de autor y evitar el plagio.

- Participación Activa: Participar activamente en clases, realizar preguntas y contribuir de manera constructiva a las discusiones académicas.

- Cumplimiento de Tareas y Plazos: Entregar las tareas y proyectos en los plazos establecidos.

- Uso Responsable de Recursos: Utilizar los recursos de la universidad, como bibliotecas y laboratorios, de manera responsable y cuidadosa.

- Respeto por los Profesores y Compañeros: Mostrar respeto hacia los profesores y compañeros.

- Comportamiento Ético y Profesional: Mantener un comportamiento ético y profesional en todo momento, tanto en el campus como fuera de él.

- Cuidado de la Salud Mental y Física: Priorizar tanto el bienestar físico como el bienestar mental.

- Seguridad en el Campus: Cumplir con las normativas de seguridad del campus, incluyendo reglas sobre el uso de laboratorios y equipo, así como medidas de seguridad en caso de emergencia.

- Integridad en los Exámenes: Seguir las reglas establecidas por la universidad durante los exámenes

## Reglas Básicas para el Comportamiento en el Transporte Público

- Respeto a las Colas: Formar una fila ordenada al abordar el transporte.

- Ceder Asientos Prioritarios: Ceder tu asiento a quienes lo necesiten.

- Volumen de la Música y Teléfonos: Mantener el volumen a un nivel que no moleste a otros pasajeros.

- Bolsos y Equipaje: Mantener el equipaje en una buena posición para evitar obstruir los pasillos y asientos.

- Basura: No dejar basura ni desechos en el transporte público

- No Fumar ni Vapear: La mayoría de los sistemas de transporte prohíben fumar o vapear en áreas designadas como estaciones o paradas.

- Respeto a los Trabajadores del Transporte: Tratar a los conductores y empleados del transporte con respeto y cortesía.

- Prioridad de Ancianos o Discapacitados: Dar prioridad a las personas mayores o con discapacidad al abordar o salir del transporte.

- Conciencia del Espacio Personal: Respetar el espacio personal de los demás.

- Seguridad: Notificar a las autoridades o al personal del transporte si ves algo sospechoso o te sientes inseguro.

# Juego De La Vida
El Juego de la Vida, creado por el matemático británico John Conway en 1970, es un ejemplo clásico de un autómata celular, un modelo matemático y computacional que simula la evolución de un conjunto de células en una cuadrícula bidimensional. Aunque es un juego simple, el Juego de la Vida es conocido por su capacidad para generar patrones sorprendentemente complejos y fascinantes a partir de reglas muy simples.

## Modelo del Juego de la Vida

El Juego de la Vida se juega en una cuadrícula bidimensional infinita, donde cada celda puede estar en uno de dos estados posibles: viva o muerta. La evolución del juego se rige por un conjunto de reglas simples:

Supervivencia: Una célula viva permanece viva en la siguiente generación si tiene 2 o 3 vecinos vivos. Si tiene menos de 2 vecinos vivos, muere por soledad, y si tiene más de 3 vecinos vivos, muere por sobrepoblación.

Nacimiento: Una célula muerta se convierte en una célula viva en la siguiente generación si tiene exactamente 3 vecinos vivos.

Estas reglas se aplican a todas las células de la cuadrícula de manera simultánea en cada generación. Como resultado, se pueden observar patrones que evolucionan a lo largo del tiempo, algunos de los cuales son estables, mientras que otros son efímeros o caóticos.

## Análisis de Resultados

El Juego de la Vida ha sido objeto de estudio en matemáticas y ciencias de la computación debido a su capacidad para generar comportamientos complejos a partir de reglas simples. Algunos de los resultados notables incluyen:

Patrones Estables: Existen patrones que permanecen inalterados a lo largo de las generaciones, como el bloque (una estructura de 4 células) o el parpadeador (una estructura que oscila entre dos estados en un ciclo de dos generaciones).

Patrones Oscilantes: Algunos patrones oscilan entre dos o más estados a lo largo de un ciclo. El "pulsar" es un ejemplo de un patrón que oscila en un ciclo de 3 generaciones.

Patrones Viajeros: Se han descubierto patrones que se desplazan a través de la cuadrícula sin cambiar su forma fundamental, como el "planeador".

Construcción de Computadoras: Se ha demostrado que el Juego de la Vida es Turing completo, lo que significa que puede utilizarse para simular una máquina de Turing y, por lo tanto, es capaz de realizar cualquier cálculo computacional.

Caos y Comportamiento Complejo: A pesar de sus reglas simples, el Juego de la Vida puede generar patrones caóticos y complejos que son difíciles de predecir a largo plazo. Estos patrones a veces se denominan "gliders" (deslizadores) y pueden interactuar de maneras inesperadas.

El Juego de la Vida de John Conway es un modelo fascinante que ilustra cómo las reglas simples pueden dar lugar a comportamientos sorprendentemente complejos en sistemas dinámicos. Ha sido un tema de estudio en matemáticas, ciencias de la computación y otras disciplinas, y ha generado una amplia variedad de patrones y estructuras interesantes que continúan siendo objeto de investigación y exploración.

# Escala De Grises

Se presenta un proceso de transformación y regeneración de imágenes que combina la conversión de imágenes a tonos de grises, el análisis de vecindades de píxeles y la predicción de colores. El proceso parte de una fotografía a color y abarca la obtención de reglas para predecir colores a partir de píxeles en escala de grises y su aplicación en la generación de imágenes a color. Posteriormente, se extiende esta metodología para colorear una imagen antigua basándose en las reglas previamente descubiertas.

El procedimiento inicia con la conversión de una imagen en color a una versión en tonos grises, reduciendo la información cromática pero manteniendo las tonalidades de luminosidad. A continuación, se aborda el análisis de cada píxel en la imagen en tonos grises y su correspondiente píxel en la imagen a color original. Mediante la aplicación de una vecindad de Moore, que involucra nueve píxeles adyacentes, se genera una regla de predicción para determinar el color del píxel en la imagen a color basándose en los colores circundantes del píxel en tonos grises.

Estas reglas de asignación de colores se recopilan y almacenan para cada valor de píxel en escala de grises, lo que resulta en un conjunto de directrices para predecir colores a partir de los niveles de gris. Utilizando estas reglas, se procede a regenerar la fotografía a color. Cada píxel en la imagen en tonos grises se colorea nuevamente siguiendo las reglas establecidas, lo que da como resultado una nueva versión a color de la imagen original.

En conjunto, este enfoque demuestra cómo la conversión de imágenes a tonos de grises y la aplicación de vecindades de píxeles pueden utilizarse para extraer reglas que guíen la generación y colorización de imágenes a color, enriqueciendo tanto las imágenes originales como las antiguas con una nueva apreciación visual y una conexión entre las tonalidades de grises y la paleta cromática.
