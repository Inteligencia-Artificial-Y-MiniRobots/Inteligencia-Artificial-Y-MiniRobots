#**Que es GPT NEO?**  
GPT-Neo es un modelo de lenguaje grande (LLM) desarrollado por EleutherAI. Es un modelo transformador preentrenado que se puede utilizar para una variedad de tareas, incluyendo generación de texto, traducción de idiomas y respuesta a preguntas.  

GPT-Neo está disponible en dos tamaños: GPT-Neo 1.3B y GPT-Neo 2.7B. GPT-Neo 1.3B tiene 1.300 millones de parámetros, mientras que GPT-Neo 2.7B tiene 2.700 millones de parámetros.  

GPT-Neo está entrenado en un conjunto de datos masivo de texto y código. El conjunto de datos incluye libros, artículos, código y otros tipos de texto.

GPT-Neo se puede utilizar para una variedad de tareas, incluyendo:

Generación de texto: GPT-Neo se puede utilizar para generar texto creativo, como poemas, historias y guiones.
Traducción de idiomas: GPT-Neo se puede utilizar para traducir idiomas, como inglés, español y francés.
Respuesta a preguntas: GPT-Neo se puede utilizar para responder a preguntas, como "¿Cuál es la capital de Francia?"
GPT-Neo es un modelo de lenguaje potente que se puede utilizar para una variedad de tareas. Es una herramienta valiosa para investigadores, desarrolladores y usuarios finales.

```python
from transformers import pipeline #pipeline en transformers es una herramienta que permite a los usuarios crear flujos de trabajo de procesamiento de lenguaje natural (NLP) personalizados.
```

# 2.Generador de Configuracion

```python
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-2.7B') #Creamos el generador de texto usando el modelo de gpt neo 2.7
```

# 3. Generando texto a partir de un Prompt
Ejecutar la linea [8] con el prompt deseado, luego la linea 9 y 10 sucesivamente para obtener el texto de salida generado con GPT NEO

```python
prompt = "El estudiante iba a la universidad" #Prompt que generara el texto
```

```python
res = generator(prompt, max_length=50, do_sample=True, temperature=0.9) #Se genera el texto a partir del prompt
```

# 4. Texto de Salida

```python
print(res[0]['generated_text']) #Imprime el texto de salida
```

```python
with open('gpttext.txt', 'w') as f:            #Se utiliza para guardar el programa en un archivo de texto
    f.writelines(res[0]['generated_text'])
```
